# Aufgabe 3 – Timetable problem mit Python constraints

## Aufgabenstellung:
>Für die Raumaufteilung am Albert-Einstein-Gymnasium bei den mündlichen Abiturprüfungen soll der Hausmeister einen Plan erstellen. Er hat folgende Informationen: Die vier Lehrer Maier, Huber, Müller und Schmid prüfen die Fächer Deutsch, Englisch, Mathe und Physik in den aufsteigend nummerierten Räumen 1, 2, 3 und 4.
> Jeder Lehrer prüft genau ein Fach in genau einem Raum.
> Außerdem weiß er folgendes über die Lehrer und ihre Fächer:
> 1. Herr Maier prüft nie in Raum 4.
> 2. Herr Müller prüft immer Deutsch.
> 3. Herr Schmid und Herr Müller prüfen nicht in benachbarten Räumen.
> 4. Frau Huber prüft Mathematik.
> 5. Physik wird immer in Raum 4 geprüft.
> 6. Deutsch und Englisch werden nicht in Raum 1 geprüft.

**Wer prüft was in welchem Raum?**

### 1. Definieren der Variablen

```python
from constraint import Problem, AllDifferentConstraint

problem = Problem() #define problem

#iterate all rooms and set with a set of all possible properties
for room_numbers in range(1,5):
    problem.addVariable(f"Lehrer{room_numbers}", ["Maier", "Huber", "Müller", "Schmid"])
    problem.addVariable(f"Fach{room_numbers}", ["Deutsch", "Englisch", "Mathe", "Physik"])
```

### 2. Hinzufügen der Constraints

```python
problem.addConstraint(
    AllDifferentConstraint(), [f"Lehrer{i}" for i in range(1, 5)])

problem.addConstraint(
    AllDifferentConstraint(), [f"Fach{i}" for i in range(1,5)])

for room_numbers in range(1,5):
    #1
    if room_numbers == 4:
        problem.addConstraint(
            lambda lehrer : lehrer != "Maier", (f"Lehrer{room_numbers}",),
        )

    #2
    problem.addConstraint(
        lambda lehrer, fach : lehrer != "Müller" or fach == "Deutsch",
                         (f"Lehrer{room_numbers}", f"Fach{room_numbers}", ),
    )

    #3
    if room_numbers < 4:
        problem.addConstraint(
            lambda lehrer_a, lehrer_b :
            not (lehrer_a == "Schmid" and lehrer_b == "Müller") and
            not (lehrer_b == "Schmid" and lehrer_a == "Müller"),
            (f"Lehrer{room_numbers}", f"Lehrer{room_numbers + 1}",),
        )

    #4
    problem.addConstraint(
        lambda lehrer, fach : lehrer != "Huber" or fach == "Mathe",
                         (f"Lehrer{room_numbers}", f"Fach{room_numbers}",),
    )

    #5
    if room_numbers == 4:
        problem.addConstraint(
            lambda fach : fach == "Physik", (f"Fach{room_numbers}",),
        )

    #6
    if room_numbers == 1:
        problem.addConstraint(
            lambda fach : fach != "Deutsch" and fach != "Englisch",
            (f"Fach{room_numbers}",),
        )

solutions = problem.getSolutions()
```

### 3 Lösung anzeigen

```python
def showSolution(solution):
    for i in range(1, 5):
        print("Raum %d" % i)
        print("--------")
        print("Lehrer: %s" % solution["Lehrer%d" % i])
        print("Fach: %s" % solution["Fach%d" % i])
        print("")

print("Found %d solution(s)!" % len(solutions))
print("")
for solution in solutions:
    showSolution(solution)
```
