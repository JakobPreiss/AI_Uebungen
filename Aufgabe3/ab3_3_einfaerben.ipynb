# Aufgabe 3 – Einfärben der Bundesländer
- Kein Bundeslandnachbar soll die gleiche Farbe besitzen

```python
from constraint import Problem

problem = Problem()

# Alle 16 Bundesländer
bundeslaender = [
    "Schleswig-Holstein",
    "Hamburg",
    "Bremen",
    "Niedersachsen",
    "Mecklenburg-Vorpommern",
    "Brandenburg",
    "Berlin",
    "Sachsen-Anhalt",
    "Nordrhein-Westfalen",
    "Hessen",
    "Thüringen",
    "Sachsen",
    "Rheinland-Pfalz",
    "Saarland",
    "Baden-Württemberg",
    "Bayern",
]

# 4 Farben genügen für jede planare Karte (Vier-Farben-Satz)
farben = ["rot", "blau", "grün", "gelb"]

for land in bundeslaender:
    problem.addVariable(land, farben)

# Nachbarschaftspaare (keine zwei Nachbarn dürfen dieselbe Farbe haben)
nachbarn = [
    ("Schleswig-Holstein", "Hamburg"),
    ("Schleswig-Holstein", "Mecklenburg-Vorpommern"),
    ("Schleswig-Holstein", "Niedersachsen"),
    ("Hamburg", "Mecklenburg-Vorpommern"),
    ("Hamburg", "Niedersachsen"),
    ("Bremen", "Niedersachsen"),
    ("Niedersachsen", "Mecklenburg-Vorpommern"),
    ("Niedersachsen", "Brandenburg"),
    ("Niedersachsen", "Sachsen-Anhalt"),
    ("Niedersachsen", "Thüringen"),
    ("Niedersachsen", "Hessen"),
    ("Niedersachsen", "Nordrhein-Westfalen"),
    ("Mecklenburg-Vorpommern", "Brandenburg"),
    ("Brandenburg", "Berlin"),
    ("Brandenburg", "Sachsen-Anhalt"),
    ("Brandenburg", "Sachsen"),
    ("Sachsen-Anhalt", "Sachsen"),
    ("Sachsen-Anhalt", "Thüringen"),
    ("Nordrhein-Westfalen", "Hessen"),
    ("Nordrhein-Westfalen", "Rheinland-Pfalz"),
    ("Hessen", "Thüringen"),
    ("Hessen", "Rheinland-Pfalz"),
    ("Hessen", "Baden-Württemberg"),
    ("Hessen", "Bayern"),
    ("Thüringen", "Sachsen"),
    ("Thüringen", "Bayern"),
    ("Sachsen", "Bayern"),
    ("Rheinland-Pfalz", "Saarland"),
    ("Rheinland-Pfalz", "Baden-Württemberg"),
    ("Baden-Württemberg", "Bayern"),
]

for land1, land2 in nachbarn:
    problem.addConstraint(lambda c1, c2: c1 != c2, (land1, land2))

loesung = problem.getSolution()

if loesung:
    print("Lösung gefunden:\n")
    for land in bundeslaender:
        print(f"  {land:<25} → {loesung[land]}")
else:
    print("Keine Lösung gefunden.")
```
